Ćwiczenie kształtuje podstawowe umiejętności posługiwania się słownikiem (`dict`) w języku Python.

:::{admonition} Polecenia do ćwiczenia
:nonumber:

Wykonuj podane kroki po kolei, w kolejnych komórkach obliczeniowych notatnika. Eksperymentuj.

1. Za pomocą `dict`, stwórz **co najmniej** 5-elementowy słownik zawierający listę zwierząt ("kot", "pies", "krowa”, "kura", "wąż", ...) i odpowiadającą im liczbę nóg.
2. Wydrukuj zawartość słownika.
3. Przetestuj, czy zawiera on określony element  (np. `kura`, albo wartość `4`).
4. Sprawdź, czy w słowniku są zwierzęta z 6 nogami.
5. Dodaj do słownika zwierzę "pająk" z liczbą nóg 6.
6. Wypisz zawartość słownika.
7. Zmień pająkowi liczbę nóg na 8.
8. Wypisz zwierzęta zapamiętane w słowniku, w kolejności rosnącej (`keys`).
9. Wypisz zbiór nóg zapamiętanych w słowniku (`values`).
10. Wypisz te zapamiętane zwierzęta, które mają 4 nogi.
11. Wypisz sumę nóg zapisanych w słowniku zwierząt.
12. Spróbuj ponownie dodać do słownika pająka z liczbą nóg 8.
13. Usuń wpis dotyczący pająka ze słownika.
14. Wypisz zwierzęta oraz liczbę nóg, w kolejności od najmniejszej liczby nóg do największej. W przypadku tej samej liczby nóg wypisz zwierzęta w kolejności rosnącej (porządek leksykograficzny).
:::
